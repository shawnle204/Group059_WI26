## Research Question

To what extent do pre-college academic indicators predict first-year college academic performance? Using a dataset containing high school GPA, standardized test scores (SAT math and verbal), coursework rigor (e.g., number of AP/IB courses), and high school context (e.g., school competitiveness), we model students’ first-year college performance. This project evaluates the relative and combined predictive power of these variables using regression and classification approaches. We further analyze whether certain preparation profiles systematically overperform or underperform relative to model expectations. This predictive modeling task aims to identify which aspects of high school success transfer most effectively to early college performance.


## Background and Prior Work

Recent national discussions around college readiness have been brought into sharp focus at UC San Diego, where reports of a growing remedial math placement rate among incoming students have gained mainstream attention. As UCSD students, we were motivated by this issue after the Academic Senate released a Senate–Administration Working Group (SAWG) report analyzing admissions practices and student academic preparedness. The report highlights a notable increase in the number of incoming students requiring remedial-level math instruction, raising concerns about the alignment between pre-college preparation and college-level expectations.<a name="cite_ref-1"></a><sup>1</sup>
 This development has prompted broader questions about how well traditional indicators of academic readiness translate into early college success.

The SAWG report attributes part of this trend to changes in UCSD’s admissions landscape, including increased enrollment of students from under-resourced and lower-funded high schools.<a name="cite_ref-1"></a><sup>1</sup>
 While the report does not claim that students from these schools are incapable of succeeding, it emphasizes that disparities in access to advanced coursework and academic resources may affect readiness for college-level mathematics. Importantly, the report suggests that commonly used admissions metrics may not fully capture differences in academic preparation, motivating further investigation into which pre-college factors meaningfully predict early college performance.

Prior research and exploratory projects have examined similar questions using simplified academic datasets. One such example is a publicly documented Kaggle project that analyzed the relationship between high school GPA, SAT scores, and first-year college GPA using linear regression.<a name="cite_ref-2"></a><sup>2</sup>
 That project found that high school GPA exhibited a slightly stronger correlation with first-year college GPA than SAT scores, suggesting that sustained academic performance may be a better indicator of early college success than standardized testing alone. The analysis also explored gender differences, observing small performance disparities despite similar SAT and high school GPA distributions.

These findings align with broader education research suggesting that no single pre-college metric fully explains college outcomes. Geiser and Santelices (2007), using University of California admissions and outcomes data, report that high school GPA is a stronger and more consistent predictor of college success than standardized test scores across multiple outcomes.<a name="cite_ref-3"></a><sup>3</sup>
 In a community college placement context, Belfield and Crosta (2012) compared placement-test-only approaches with transcript-based predictors and found that using high school performance indicators can improve placement decisions and reduce misclassification.<a name="cite_ref-4"></a><sup>4</sup>
 More recent work by Allensworth and Clark (2020), which uses hierarchical modeling across high schools, shows that GPA remains a relatively stable predictor while test-score predictive power varies more across school contexts.<a name="cite_ref-5"></a><sup>5</sup>
 Together, these methods-based findings suggest that model performance depends on both variable choice and institutional context, not just the algorithm itself.

Building on both institutional reporting and prior data-driven projects, our work seeks to extend this line of inquiry by examining a broader set of pre-college academic indicators and their relationship to early college performance. In addition to SAT and high school GPA, prior studies motivate including variables tied to coursework rigor and educational context (for example, AP/IB exposure, first-generation status, and school-resource context) because these factors may mediate how test scores translate into first-year outcomes.<a name="cite_ref-4"></a><sup>4</sup><a name="cite_ref-5"></a><sup>,5</sup>
 By situating our analysis within the context of the UCSD remedial math discussion and grounding our variable selection in existing methods-focused literature, this project aims to better understand the practical limits of commonly used readiness metrics and identify which aspects of high school preparation are most strongly associated with first-year college success.

# Potential Predictor Columns
### Student Context
- Curriculum: 'RIGOR', 

> Curriculum pertains to the intensity and composition of their coursework. Values indicate whether students completed a college-bound core curriculum, a college-bound core curriculum without computer science, a standard core curriculum without computer science, a minimal core curriculum, or other curricular patterns. This variable is indicative of coursework rigor and academic preparation prior to college.

- Grades (9-12, Academic Track, and Weighted): 'HS_GPA_9TH', 'HS_GPA_10TH', 'HS_GPA_11TH', 'HS_GPA_12TH', 'HS_GPA_ACAD','HS_GPA_WGT', 

> Grades (gpa) pertains to the level of success a student was able to achieve throughout their high school career. Grade specific gpas represent annual academic performance on a standard gpa scale, allowing analysis of trends over time. The academic-track gpa (X3TGPAACAD) reflects performance specifically in college-preparatory courses, while the weighted GPA (X3TGPAWGT) adjusts for course difficulty by giving additional weight to honors, AP, or IB classes. Together, these GPA metrics provide a comprehensive measure of sustained academic performance and/or academic growth through high school.

- Time Spent Studying: 'HOURS_HOMEWORK'

> These columns give the average time each student spent studying for math, science, and other homework in total for each school day in high school. This can reveal how much time a student committed to working in school, which can reflect on their grades. A student who has dedicated more time into school is expected to be more prepared for higher education than students do not. This column is feature engineered through combining the specific study times for math, science, and other, where the original columns were dropped in favor of total study time.

- AP/IB Credit: 'TOTAL_AP_CREDITS'
> Since AP/IB credit is meant to be college level courses in high school that allows the student to get college credit if they do well enough on an objective exam that is given at the end of the course content. If they do get college credit, this can mean that they are more prepared for college-level course work than others who have not.

### School Context
- School Avg SAT Per School: 'SAT_SCHOOL_AVG'

> These include the average SAT Math and Reading scores for each student’s high school. These can be indicative of school competitiveness and peer academic ability. - School Avg ACT Per School: 'ACT_SCHOOL_AVG'

> The dataset also includes school average ACT performance. Similar to the school average SAT variables, these metrics capture the academic environment and competitiveness of each student's high school. Schools with higher average ACT scores might provide stronger academic preparation and academically engaged peer groups, which may indirectly affect students’ readiness for college coursework.

- Standardized School Score (by Administrators): 'SCHOOL_PROBLEM_SCALE', 'SCHOOL_CLIMATE_SCALE'

> The dataset includes a standardized school score reported by administrators. This measure the school academic and social environment which may play a factor into student's academic performance. These social factors may impact student performance and their post-secondary studies through influence and traits taken by other students or social contexts. However, since the score is reported by school administrators, there are potential concerns regarding false-reported scores, which indicate bias in favor of the school.

- Standardized School Score (by students): 'STUDENT_SCHOOL_MOTIVATION'

> The dataset also includes a standardized school score reported by the students themselves. This indicates the student's own perspectives on the school environment rather than the eyes of the school administrators.

#### Freshman GPA vs High School GPA
This LM plot is used to show whether or not there is some correlation in high school GPA and college freshman GPA for each student. The plot illustrates a general upward slope that suggests there is some correlation, but the points are very spread out, so the correlation is low. There are instances of some students received a high school GPA, but their freshman GPA shows close to a 0, which indicates some external circumstances such as dropping out. #### AP/IB Credit vs. Ratio of Credits Earned
This scatterplot shows how much AP/IB credit the student earned in high school, which is meant to be a college-level course, versis the ratio of credits earned over credits attempted. This shows the ratio of classes the student passed, similar to how AP/IB credit works. AP/IB credit is earned when a student passes an exam that is cummulative of the contents of the class. Whether or not they receive credit should be similar to whether or not they are going to pass an equivalent class in college. This means that the ratio in which they are earning credit should correlate to how much AP/IB credit they earned. In this case, a ratio of 1 should be the max, however, there seems to be some error where some students are earning twice as much credit as they have attempted. Additionally, it seems that most students did not have AP or IB credit coming into college, so there is not as much data to work off of for this variable.

```python
credits_df = comb_df[['TOTAL_AP_CREDITS', 'FRESHMAN_GPA']]
credits_df = credits_df.assign(ratio = (comb_df['FRESHMAN_CREDITS_EARNED'] / comb_df['FRESHMAN_CREDITS_ATTEMPTED']))
credits_df = credits_df.dropna()
credits_df = credits_df[(credits_df['ratio'] >= 0) &
                        (credits_df['ratio'] <= 8) &
                        (credits_df['TOTAL_AP_CREDITS'] >= 0)]
sns.scatterplot(data = credits_df, x = 'TOTAL_AP_CREDITS', y = 'ratio')
plt.xlabel('AP and IB Credit')
plt.ylabel('Ratio of Credits Earned Over Credits Attempted')
plt.title('AP/IB Credit VS. Ratio of Credits Earned')
plt.tight_layout()
plt.show()
```
